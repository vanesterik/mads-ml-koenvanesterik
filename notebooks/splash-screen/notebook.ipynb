```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Setup the data
# We create a grid of X and Y values
grid_range = 3
resolution = 40  # Higher number = more grid lines (denser mesh)
x = np.linspace(-grid_range, grid_range, resolution)
y = np.linspace(-grid_range, grid_range, resolution)
X, Y = np.meshgrid(x, y)

# 2. Define the Z-axis (The Funnel Shape)
# Calculate distance from center (radius)
R = np.sqrt(X**2 + Y**2)

# Create the funnel.
# We add a small value (0.1) to avoid division by zero at the center
Z = -1 / (R**2 + 0.1)

# 3. Initialize the Plot
fig = plt.figure(figsize=(16, 9), facecolor="black")  # Black window background
ax = fig.add_subplot(111, projection="3d")

# 4. Styling the Plot Area
ax.set_facecolor("black")  # Black plot background
ax.grid(False)  # Turn off the default gray grid lines
ax.axis("off")  # Turn off the axes (x, y, z numbers)

# 5. Plot the Wireframe
# rstride and cstride control how many lines are drawn (step size)
# linewidth makes the lines delicate like the reference image
ax.plot_surface(
    X,
    Y,
    Z,
    rstride=1,
    cstride=1,
    facecolor="#000000",
    edgecolor="#FFFFFF",
    linewidth=1,
    shade=False,
)

# 6. Adjust View Angle
# elev = elevation (up/down), azim = azimuth (rotation)
ax.view_init(elev=18, azim=45)

# ZOOM SETTINGS
zoom_window = 1.5  # The original data goes from -3 to 3.
# Setting this to 1.5 crops the view to the center.

ax.set_xlim(-zoom_window, zoom_window)
ax.set_ylim(-zoom_window, zoom_window)

# NOTE: When you zoom in on X and Y, you might want to limit Z
# as well so the funnel doesn't look infinitely deep or stretched.
ax.set_zlim(-21, 0)

# Optional: Tight layout to remove white margins
plt.tight_layout()

plt.show()
```
